```
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import base64

# login credentials of user
user_name = "nikhil"
user_pass = "2002"

print("Smart Infrastructure Security System")

user = input("Enter user_name: ")
pwd = input("Enter Password: ")

if user == user_name and pwd == user_pass:
    print("\nLogin Successful- Welcome User ")

    # this part encrypts data using AES
    data = input("\nEnter sensor data to encrypt: ").encode()

    key = get_random_bytes(16)  # 16 bytes key for AES
    cipher = AES.new(key, AES.MODE_EAX)

    ciphertext, tag = cipher.encrypt_and_digest(data)

    print("\nEncrypted Data:", base64.b64encode(ciphertext).decode())

    # this part decrypts data using DES
    cipher_dec = AES.new(key, AES.MODE_EAX, nonce=cipher.nonce)
    decrypted_data = cipher_dec.decrypt(ciphertext)

    print("Decrypted Data:", decrypted_data.decode())

else:
    print("\nAccess not Authorized - access not allowed ")
```
